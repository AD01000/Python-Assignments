```python
#1.Program to read the details of a student & display it.
name = input('Enter the name:')         #Read inputs.
roll = input('Enter the roll number:')
mark = input('Enter the mark:')
print('Name:',name)                    #Display details.
print('Roll No:',roll)
print('Mark:',mark)
```

```python
#2.Program to convert Celcius(C) to Fahrenheit(F) & inversely.
t = str(input('Enter the unit of Temperature:'))   #Read the Unit of temperature
if(t =='F'):
  a = float(input('Temperature in Fahrenheit:'))   #Read the Temperature in F.
  ce = (a-32)*(5/9)                                      
  print('The temperature in Celsius:', ce)         #Output in Celsius.
  
else:   
  a = float(input('Temperature in Celcius:'))      #Read the Temperature in C.                                    
  fa=(9*a/5)+32
  print('The temperature in Fahrenheit:',fa)       #Output in Fahrenheit.   
  
```

```python
#3.Program to read two numbers and display its product.
b = int(input('Enter the first number:'))   #Read the numbers.
c = int(input('Enter the second number:'))
d = b*c                                     #Calculate the Product.
print('The product of the numbers:',d)      #Output.
```

```python
#4.Program to add a characters to string.
l ="Learn Python by Developing a Website"                #Given String.
l = l[:5]+"**"+l[6:12]+"**"+l[12:29]+"**"+l[29:37]+"**"  #Using + to add character to the string.
print(l)                                                 #Output.Updated string.
```

```python
#5.Program to read the radius of a circle & display its area.
import math                              #Importing math module for an accurate output.
rad = int(input('Enter the radius :'))   #Read the radius.
area = math.pi*rad**2                    #Calculate the area using math module.
print('The Area of the circle:',area)    #Output.
```

```python
#6.Program to insert a number to any position in a list.
list1 = [1,3,2,4,7,5,9,23,54,78,21]        #List.
n = int(input('Enter the number:'))        #Read the Number to be inserted.
p = int(input('Enter the position:'))      #Read the Position.
list1.insert(p,n)                          #Using insert method to add the number to read position.
print(list1)                               #Output.Updated list.
```

```python
#7.Program to Rename the key city in a dictionary.
sample_dict = {"name":"John","age":5,"salary":8000,"city":"New York"} #Given Dictionary.
i = str(input('Enter the Location name:'))                            #Read the Location to be inserted.
sample_dict["city"] = i                                               #Rename the key city.
print(sample_dict)                                                    #Output.Updated Dictionary.
```

```python
#8.Program to change a value of a key in a Dictionary.
sample_dict = {'emp1':{'name':'John','salary':7500},'emp2':{'name':'Emma','salary':8000},'emp3':{'name':'Brad','salary':500}} #Given Dictionary.
sample_dict['emp3']['salary'] = 8500                                                                                           #Update the salary.
print(sample_dict)                                                                                                            #Output.Updated Dictionary.
```
